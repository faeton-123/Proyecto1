### (5.4) 
Basado en la gráfica y el coeficiente de correlación de cada par x,y elegir las 2 variables con más potencial predictivo es decir las 2 variables que presentan mayor correlación entre dicha variable y la variable dependiente.

En la seccion 5.1 al calcular los factores de correlación, los datos que mejor correlación tiene son las columnas: OverallQual y 1stFlrSF, respecto a la columna SalePrice


```python
def entreno_regresion(x, y, epochs, imprimir_error_cada, alpha):
    # Crear matriz de características
    X = np.column_stack((y, np.ones(len(y))))
    
    # Inicializar parámetros del modelo
    beta = np.zeros(2)
    
    # Almacenar errores para su posterior visualización
    errors = []
    
    # Entrenar modelo
    for i in range(epochs):
        # Calcular predicciones
        y_pred = np.dot(X, beta)
        
        # Calcular error del modelo
        error = np.mean((y_pred - y) ** 2)
        errors.append(error)
        
        # Actualizar parámetros del modelo
        beta -= alpha * np.dot(X.T, y_pred - x) / len(x)
        
        # Imprimir error cada imprimir_error_cada iteraciones
        if i % imprimir_error_cada == 0:
            print("Iteración:", i, "Error:", error)
    
    return beta, errors
```

```python

#x = data_80['SalePrice: ']
#y = data_80['OverallQual: ']
#y = data_80['1stFlrSF: ']
#y = data_80['TotRmsAbvGrd: ']
#y = data_80['YearBuilt: ']
#y = data_80['LotFrontage: ']

# llamar a la función train_linear_regression()
entreno_regresion(data_80['SalePrice: '], data_80['TotRmsAbvGrd: '], epochs=20, imprimir_error_cada=1, alpha=0.5)
```
